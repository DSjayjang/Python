```python
# Data Load
from sklearn import datasets

iris = datasets.load_iris()
print(iris)
```

```python
from sklearn.metrics import confusion_matrix

y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]

# 대각 원소는 실제값과 예측값이 일치하는 경우
# 행과 열은 차례대로 0, 1, 2
# 행이 y_true
# 열이 y_pred
confusion_matrix(y_true, y_pred)
```

```python
# 데이터 불러오기
from sklearn import datasets

raw_wine = datasets.load_wine()
```

```python
raw_wine
```

```python
# feature, target
X = raw_wine.data
y = raw_wine.target

print(X.shape)
print(y.shape)
```

```python
# 트레이닝 / 테스트 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# 데이터 scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaling = scaler.transform(X_train)
X_test_scaling = scaler.transform(X_test)
```

```python
from sklearn import svm
```

```python
svm_model = svm.SVC(kernel='linear', random_state=0)
```

```python
svm_model.fit(X_train_scaling, y_train)
```

```python
y_pred =svm_model.predict(X_test_scaling)
y_pred 
```

```python
# 평가
# 정확도
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

# confusion matrix
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

```python
# 분류 리포트
from sklearn.metrics import classification_report

class_report = classification_report(y_test, y_pred)
print(class_report)
```
