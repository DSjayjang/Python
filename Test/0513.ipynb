```python
N = int(input())
A = [list(map(int, input().split())) for _ in range(N)]
R = int(input())
C = int(input())
K = int(input())
```

```python
li = []
sum= 0 

for i in range(R-1, N):
    for j in range(C-1, N):
        li.append(A[i][j])

for i in li:
    if (i % K != 0):
        sum = sum + i
print(sum)
```
