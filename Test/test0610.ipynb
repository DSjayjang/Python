# 제1유형

```python
import pandas as pd
import numpy as np
```

```python
# dataset
# 주어지는 데이터는 153일 동안 공기의 질을 측정한 데이터셋이다.
data = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv')
print(data)
```

```python
# 1. airquality 데이터에 대해서 결측치가 가장 많은 변수를 찾아서 해당 결측치를 0으로 대치하고,
# 결측치를 제외한 평균과 0으로 대치한 후의 평균과의 차이를 구하시오.
print(data.shape)
print(data.info())

# 결측치 개수 확인
print('결측치 개수:', data['Ozone'].isnull().sum())

mean_ozone_before = np.mean(data['Ozone'])
print('mean_ozone_before:', mean_ozone_before)

data['Ozone_impute'] = data['Ozone'].fillna(0)

mean_ozone_after = np.mean(data['Ozone_impute'])
print('mean_ozone_after:', mean_ozone_after)

print('차이:', mean_ozone_before - mean_ozone_after)
```

```python
# Wind 변수에 대해 Min-Max 정규화를 수행한 후 평균값과 Z 정규화를 수행한 후 평균 값의 차이를 구하시오.
# Min-Max 정규화
X = data[['Wind']]

from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
data['Wind_minmax_scaled'] = scaler1.fit_transform(X)
mean_wind_minmax_scaled = np.mean(data['Wind_minmax_scaled'])

# Z 정규화
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
data['Wind_Z_scaled'] = scaler2.fit_transform(X)
mean_wind_Z_scaled = np.mean(data['Wind_Z_scaled'])

print('차이:', mean_wind_minmax_scaled - mean_wind_Z_scaled)
```

```python
# 월별(5월~9월) 평균 기온을 구하시오.
data.groupby(['Month'])['Temp'].mean()
```

# 제2유형

```python
# 은행에서 수집한 고객 5,000명의 금융정보에 따른 대출여부가 들어있는 참조데이터를 이용하여 대출여부를 분류하는 가장 최적의 이웃의 크기값(k)을 구하고, 이때 분류정확도를 산출하시오.
# (단, 참조데이터는 7:3의 비율로 트레이닝 데이터와 테스트 데이터로 구분하고, 트레이닝 데이터와 테스트 데이터의 대출여부(y)의 비율도 유지한다. 또한 normalizer를 사용하여 스케일링을 진행한다.)
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
```

```python
# dataset
# 은행 고객 5,000명에 대한 데이터로 인구통계정보(연령, 소득 등), 고객과의 관계(담보, 증권계좌 등), 개인대출 등의 금융정보를 포함한다.
data = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/Bank_Personal_Loan_Modeling.csv')
print(data)
```

```python
# EDA
print(data.shape)
print(data.info())
print(data.describe())
```

```python
# Data cleansing
df_new = data.drop(['ID', 'ZIP Code'], axis=1)
df_new
```

```python
X = df_new.drop(['Personal Loan'], axis=1)
y = df_new['Personal Loan']
```

```python
# 데이터 전처리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=1234)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 표준화
from sklearn.preprocessing import Normalizer
scaler = Normalizer()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# KNN
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []

# k는 1에서 25로 세팅
neighbors_settings = range(1, 25)

# KNeighborsClassifier() 함수에 k개수를 입력하여 모델을 만들고 트레이닝 데이터와 테스트 데이터를 적용
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train_scaled, y_train)

    # 훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train_scaled, y_train))

    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test_scaled, y_test))
```

```python
# 최적의 이웃 수 산출
# k=5일 때 테스트 정확도가 92.6%로 가장 높다.
test_accuracy
```

# 제3유형

```python
# 측정한 온도(Temp)의 평균값이 평균온도인 77과 일치하는지 t-검정을 이용하여 확인하시오.
```

```python
import pandas as pd
import numpy as np
```

```python
# data
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv')
print(df)
```

```python
# 1. 표본평균 X를 소수점 둘째자리까지 반올림하여 구하시오.
X = df['Temp']
mean_X = np.round(np.mean(X), 2)
print('평균:', mean_X)
```

```python
# 2. 가설검정을 위한 검정통계량을 소수점 둘째자리까지 반올림하여 구하시오.
from scipy import stats

t_statistic, p_value = stats.ttest_1samp(X, popmean=77)
print(round(t_statistic, 2))
print(round(p_value, 4))
```

```python
# 3. 통계량에 대한 p-value 값을 소수점 넷째자리까지 반올림하여 구하고, 유의수준 0.05 내에서 결과를 논하시오. (채택과 기각 중 선택)
p = np.round(p_value, 4)
print(p)

if p < 0.05:
    print('기각')
else:
    print('채택')
```
