```python
import numpy as np
from scipy.stats import norm
```

```python
# data
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])
```

```python
# 모집단의 평균과 분산 (표준편차)
population_mean = 28
population_variance = 4

# 유의수준 설정
alpha = 0.05

# 단측 또는 양측 검정 선택
alternative = 'two-sided'

# 표본평균 계산
sample_mean = np.mean(data)

# 표본 크기
n = len(data)
```

```python
# 검정통계량 계산 (Z-검정)
test_statistic = (sample_mean - population_mean) / np.sqrt(population_variance/n)
print(test_statistic)

# 임계값 계산 (Z-검정, 양측 검정)
critical_value = norm.ppf(1 - alpha / 2)
print(critical_value)

# p-value 계산 (양측 검정)
p_value = 2 * (1 - norm.cdf(abs(test_statistic)))
print(p_value)
```

```python
# 결과
print('표본평균:', sample_mean)
print('검정통계량:', test_statistic)
print('임계값:', critical_value)
print('p-값:', p_value)

# 결과 해석
if abs(test_statistic) > critical_value:
    print('귀무가설 기각, 표본평균이 모집단 평균과 다름')
else:
    print('귀무가설 기각하지 못함, 표본평균과 모집단 평균과 차이가 없을 가능성 있음')
```

```python
from scipy import stats

# data
data = [2.4, 2.7, 3.1, 3.2, 3.5, 3.7, 3.9, 4.1, 4.2]

# Shapiro-Wilk 테스트
statistic, p_value = stats.shapiro(data)

# 결과
alpha = 0.05
print('Shapiro-Wilk 테스트 통계량:', statistic)
print('p-값 (two-tailed):', p_value)

if p_value < alpha:
    print('귀무가설 기각, 데이터는 정규분포를 따르지 않을 가능성 있음')
else:
    print('귀무가설 기각하지 못함. 데이터는 정규분포를 따를 가능성이 높음')
```

```python
# data
data_man = [117, 108, 105, 89, 101, 93, 96, 108, 108, 94, 93, 112, 92, 91, 100, 96, 120, 86, 96, 95]
data_woman = [121, 101, 102, 114, 103, 105, 101, 131, 96, 109, 109, 113, 115, 94, 108, 96, 110, 112, 120, 100]

# 평균
mean_man = np.mean(data_man)
mean_woman = np.mean(data_woman)
print('남자 데이터 평균:', mean_man)
print('여자 데이터 평균:', mean_woman)

# Shapiro-Wilk 정규성 테스트
statistic_man, p_value_man = stats.shapiro(data_man)
statistic_woman, p_value_woman = stats.shapiro(data_woman)

# 결과
alpha = 0.05

print('남자 데이터 Shapiro-Wilk 테스트 통계량:', statistic_man)
print('남자 데이터 p-값 (two-tailed):', p_value_man)

if p_value_man < alpha:
    print('귀무가설 기각, 남자 데이터는 정규분포를 따르지 않을 가능성 있음')
else:
    print('귀무가설 기각하지 못함. 남자 데이터는 정규분포를 따를 가능성이 높음')

print('여자 데이터 Shapiro-Wilk 테스트 통계량:', statistic_woman)
print('여자 데이터 p-값 (two-tailed):', p_value_woman)

if p_value_woman < alpha:
    print('귀무가설 기각, 여자 데이터는 정규분포를 따르지 않을 가능성 있음')
else:
    print('귀무가설 기각하지 못함. 여자 데이터는 정규분포를 따를 가능성이 높음')
```

```python
import numpy as np
from scipy import stats

# data
data = np.array([14.2, 15.1, 14.8, 15.4, 14.9, 15.2, 14.6, 15.0])

# 가설 설정
# H0 : 모분산은 1.0과 같다.
# H1 : 모분산은 1.0과 다르다.

population_variance = 1.0 # 모분산

# 검정통계량 = (n-1) * 표본분산 / 모분산
sample_variance = np.var(data, ddof=1)
n = len(data)
chi2_stat = (n-1) * sample_variance / population_variance

# 카이제곱 분포의 자유도
degrees_of_freedom = n-1

# p-값 계산
p_value = 1 - stats.chi2.cdf(chi2_stat, degrees_of_freedom)

# 결과
print('Chi-Square Statistic:', chi2_stat)
print('p-value:', p_value)

# 유의수준 0.05로 가설 검정
alpha = 0.05
if p_value < alpha:
    print('귀무가설 기각. 모분산은 1.0과 다름')
else:
    print('귀무가설 기각 X. 모분산은 1.0과 같을 가능성 있음.')
```

```python
import numpy as np
np.random.seed(123)

# data 생성
sample1 = np.random.normal(loc=5, scale=2, size=50)
sample2 = np.random.normal(loc=5, scale=3, size=50)
print(sample1)
print(sample2)
```

```python
from scipy import stats

# 두 표본의 분산 계산
variance_sample1 = np.var(sample1, ddof=1)
variance_sample2 = np.var(sample2, ddof=1)

# F-검정
f_statistic = variance_sample1 / variance_sample2
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

print(f'표본 1의 분산: {variance_sample1:.4f}')
print(f'표본 2의 분산: {variance_sample2:.4f}')
print(f'F-통계량: {f_statistic:.4f}')

alpha = 0.05
if p_value < alpha:
    print(f'p-value: {p_value:.4f} < {alpha}, 귀무가설 기각. 두 표본의 분산은 다름.')
else:
    print(f'p-value: {p_value:.4f} >= {alpha}, 귀무가설 기각 X. 두 표본의 분산은 같을 수 있음.')
```

```python
import numpy as np
from scipy import stats

# data
data_matrix = np.array([[50, 75, 125, 175],
                        [90, 30, 45, 10]])

# 카이제곱 독립성 검정
chi2, p, df, expected = stats.chi2_contingency(data_matrix)

# 결과
print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', df)
print('Expected frequencies table:')
print(expected)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# 타이타닉 데이터셋
titanic_data = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv')
print(titanic_data.head(5))
```

```python
# 교차표 생성: 성별(Sex)과 생존 여부(Survived)를 기반
cross_table = pd.crosstab(titanic_data['Sex'], titanic_data['Survived'])
print(cross_table)

# 카이제곱 독립성 검정
chi2, p, df, expected = chi2_contingency(cross_table)

# 결과
print('카이제곱 통계량:', chi2)
print('p-value:', p)
print('자유도(df)', df)
print('기대값(Expected Frequencies)')
print(expected)
```

```python
# 나이를 그룹으로 나누는 함수 정의
def categorize_age(age):
    if age < 20:
        return '10대'
    elif age < 30:
        return '20대'
    elif age < 40:
        return '30대'
    elif age < 50:
        return '40대'
    elif age < 60:
        return '50대'
    else:
        return '60대 이상'

# 나이가 NaN인 경우 해당 열의 평균값으로 대체
average_age = titanic_data['Age'].mean()
titanic_data['Age'] = titanic_data['Age'].apply(lambda x: average_age if np.isnan(x) else x)

# apply() 함수로 새로운 열 생성
titanic_data['AgeGroup'] = titanic_data['Age'].apply(categorize_age)

# 결과
print(titanic_data[['Age', 'AgeGroup']])
```

```python
# 교차표 생성: 연령대(AgeGroup)와 생존 여부(Survived)를 기반
cross_table = pd.crosstab(titanic_data['AgeGroup'], titanic_data['Survived'])

# 카이제곱 독립성 검정
chi2, p, df, expected = chi2_contingency(cross_table)

# 결과
print('카이제곱 통계량:', chi2)
print('p-value:', p)
print('자유도(df)', df)
print('기대값(Expected Frequencies)')
print(expected)
```
