```python
import numpy as np
from scipy.stats import norm
```

```python
# data
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])
```

```python
# 모집단의 평균과 분산 (표준편차)
population_mean = 28
population_variance = 4

# 유의수준 설정
alpha = 0.05

# 단측 또는 양측 검정 선택
alternative = 'two-sided'

# 표본평균 계산
sample_mean = np.mean(data)

# 표본 크기
n = len(data)
```

```python
# 검정통계량 계산 (Z-검정)
test_statistic = (sample_mean - population_mean) / np.sqrt(population_variance/n)
print(test_statistic)

# 임계값 계산 (Z-검정, 양측 검정)
critical_value = norm.ppf(1 - alpha / 2)
print(critical_value)

# p-value 계산 (양측 검정)
p_value = 2 * (1 - norm.cdf(abs(test_statistic)))
print(p_value)
```

```python
1- norm.cdf(abs(test_statistic))
```

```python
# 결과
print('표본평균:', sample_mean)
print('검정통계량:', test_statistic)
print('임계값:', critical_value)
print('p-값:', p_value)

# 결과 해석
if abs(test_statistic) > critical_value:
    print('귀무가설 기각, 표본평균이 모집단 평균과 다름')
else:
    print('귀무가설 기각하지 못함, 표본평균과 모집단 평균과 차이가 없을 가능성 있음')
```

```python
# data
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])
```

```python
# 유의수준 설정
alpha = 0.05

# 단측 또는 양측 검정 선택
alternative = 'two-sided'
```

```python

from scipy import stats
# t-검정 수행
t_statistic, p_value = stats.ttest_1samp(data, popmean=0, alternative=alternative)

# 95% 신뢰구간 계산
confidence_interval = stats.t.interval(confidence=1-alpha, df=len(data) - 1, loc=np.mean(data), scale=stats.sem(data))
```

```python
t_statistic, p_value = stats.ttest_1samp(data, popmean=0, alternative=alternative)

print(t_statistic, p_value)
```

```python
# 결과
print('t-통계량:', t_statistic)
print('p-값:', p_value)
print('95% 신뢰구간:', confidence_interval)

# 결과 해석
if p_value < alpha:
    print('귀무가설 기각, 표본평균이 모집단 평균과 다름')
else:
    print('귀무가설 기각하지 못함, 표본평균과 모집단 평균과 차이가 없을 가능성 있음')
```

```python
import pandas as pd
import numpy as np
from scipy import stats

mtcars = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/mtcars.csv')
```

```python
print(mtcars.head(5))
```

```python
# mpg 열의 표본평균 계산 및 반올림
mpg_mean = round(np.mean(mtcars['mpg']), 4)
print('mpg_mean:', mpg_mean)
```

```python
# mpg 열의 표본분산 계산 및 반올림
mpg_var = round(np.var(mtcars['mpg'], ddof=1), 4)
print('mpg_var:', mpg_var)
```

```python
# 단일표본 t-검정 수행, mpg 열의 평균이 20인지 검정
t_stat, p_value = stats.ttest_1samp(mtcars['mpg'], popmean=20)
print('단일표본 t-검정 결과:')
print('t-통계량:', t_stat)
print('p-값:', p_value)
```

```python
# mpg 평균의 95% 신뢰구간 계산 및 출력
conf_interval = stats.t.interval(confidence=0.95, df=len(mtcars['mpg'])-1, loc=np.mean(mtcars['mpg']), scale=stats.sem(mtcars['mpg']))
```

```python
print('mpg 평균의 95% 신뢰구간:', round(conf_interval[0], 4), '-', round(conf_interval[1], 4))
```
