# input()

```python
a = input('입력하세요:')
print('너의 입력은:', a)
```

```python
a = int(input('정수 입력하세요:'))
print('너의 입력은:', a)
```

```python
a = input('입력하세요:')

print('너의 입력은:', a)

print('type(a):', type(a))
```

# input().split()

```python
a, b, c = input().split()
print(a)
print(b)
print(c)
```

```python
num1, num2 = map(int, input().split())
print('두 수의 합은', num1 + num2, '입니다.')
```
