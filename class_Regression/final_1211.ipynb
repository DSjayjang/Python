```python
# 7
# (a) relative risk
print('relative risk:', (72/(72+28))/(60/(60+40)))

# (b) odds
odds_1 = 0.72/0.28
odds_2 = 0.60/0.40
print('odds 1:', odds_1)
print('odds 2:', odds_1)
odds_ratio =  odds_1/odds_2
print('odds ratio:', odds_ratio)

# (c)
log_odds_ratio = np.log(odds_ratio)
z = 1.96
se = np.sqrt(1/72+1/28+1/60+1/40)

CI_minus = log_odds_ratio - z * se
CI_plus = log_odds_ratio + z * se

print('\n')
print('log_odds_ratio:', log_odds_ratio)
print('SE:', se)
print('CI:', CI_minus, CI_plus)
print('오즈비 신뢰구간:', np.exp(CI_minus), np.exp(CI_plus))

print('\n')
print('CI 대충:', 0.54-1.96*0.30,  0.54+1.96*0.30)
print('오즈비 신뢰구간 대충:', np.exp(-0.048), np.exp(1.13))
```
